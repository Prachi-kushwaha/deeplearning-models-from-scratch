```python
import torch.nn as nn
import torch
import torch.nn.functional as F

class MySimpleANN:
    def __init__(self, num_features):
        self.W1 = torch.randn(num_features, 8, requires_grad=True)
        self.b1 = torch.zeros(8, requires_grad=True)

        self.W2 = torch.randn(8, 5, requires_grad=True)
        self.b2 = torch.zeros(5, requires_grad=True)

        self.W3 = torch.randn(5, 3, requires_grad=True)
        self.b3 = torch.zeros(3, requires_grad=True)

        self.W4 = torch.randn(3, 1, requires_grad=True)
        self.b4 = torch.zeros(1, requires_grad=True)

    def forward(self, x):
        x = F.relu(x @ self.W1 + self.b1)
        x = F.relu(x @ self.W2 + self.b2)
        x = F.relu(x @ self.W3 + self.b3)
        return x @ self.W4 + self.b4

    def bce_with_logits_loss(self, logits, y):
        loss = torch.clamp(logits, min=0) \
               - logits * y \
               + torch.log1p(torch.exp(-torch.abs(logits)))
        return loss.mean()

    def parameters(self):
        return [
            self.W1, self.b1,
            self.W2, self.b2,
            self.W3, self.b3,
            self.W4, self.b4
        ]

```
