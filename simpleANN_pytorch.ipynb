```python
import torch
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv("Churn_Modelling.csv")
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,:-1], df.iloc[:,-1:], test_size=0.3)
```

```python
X_train.head()
X_train.shape
```

```python
y_train.head()
y_test.shape
```

```python
ohe = OneHotEncoder(drop='first',sparse_output=False)
X_train_new_geography = ohe.fit_transform(X_train[['Geography']])
X_test_new_geography = ohe.transform(X_test[['Geography']])
```

```python
X_train_new_geography.shape
```

```python
X_train_droped = X_train.drop(columns=['Geography']).values
X_test_droped = X_test.drop(columns=['Geography']).values
```

```python
X_train_transformed = np.concatenate((X_train_new_geography,X_train_droped),axis=1)
X_test_transformed = np.concatenate((X_test_new_geography,X_test_droped),axis=1)

```

```python
from sklearn.preprocessing import StandardScaler
se = StandardScaler()
X_train_new = se.fit_transform(X_train_transformed)
X_test_new = se.transform(X_test_transformed)
```

```python
import pandas as pd

X_test_transformed = pd.DataFrame(
    X_test_new # optional but recommended
)

X_test_transformed.head()
```

```python
features = [
    'Geography_Germany',
    'Geography_Spain',
    'CustomerId',
    'CreditScore',
    'Gender',
    'Age',
    'Tenure',
    'Balance',
    'NumOfProducts',
    'HasCrCard',
    'IsActiveMember',
    'EstimatedSalary'
]
```

```python
X_test_transformed = pd.DataFrame(X_test_transformed)
X_test_transformed.columns = features
X_test_transformed.head()
```

```python
X_train_tensor = torch.from_numpy(X_train_transformed.astype(np.float32))
X_test_tensor = torch.from_numpy(X_test_transformed.to_numpy().astype(np.float32))
y_train_tensor = torch.from_numpy(y_train.to_numpy().astype(np.float32))
y_test_tensor = torch.from_numpy(y_test.to_numpy().astype(np.float32))
```

```python
import torch.nn as nn
import torch.nn.functional as F

class MySimpleANN(nn.Module):

    def __init__(self, num_features):
        super().__init__()
        self.fc1 = nn.Linear(num_features, 8)
        self.fc2 = nn.Linear(8, 5)
        self.fc3 = nn.Linear(5, 3)
        self.fc4 = nn.Linear(3, 1)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        return self.fc4(x)   # NO sigmoid

```

```python
learning_rate = 0.01
epochs = 30
```

```python
loss_function = nn.BCEWithLogitsLoss()
```

#traning pipeline

```python
# y_train_tensor.view(-1,1)
```

```python

model = MySimpleANN(X_train_tensor.shape[1])

optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

for epoch in range(epochs):
  y_pred = model(X_train_tensor)
  loss = loss_function(y_pred, y_train_tensor.view(-1,1))
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
  print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
```

```python
# model evaluation
with torch.no_grad():
  y_pred = model.forward(X_test_tensor)
  y_pred = (y_pred > 0.5).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
